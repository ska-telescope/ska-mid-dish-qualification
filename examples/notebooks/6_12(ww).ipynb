```python
#Import of Python available libraries
import sculib

mpi = sculib.scu()
mpi.ip = '10.96.64.10'
#sim = sculib.scu()
#sim.ip = '10.96.254.76'
```

```python
def mpi_t686(scu):
    print('Test 6.12: Maximum Slew Rate Tests')
    
    # Common setup
    config_name = 'hn_servo_performance'
###    config_name = 'HN_TILT_TEST'
    
    loop1 = 3 #^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^ 
    tcycle = 75 #75s for El and Combined;150s for Az tests<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<
    
    az_speed = 3
    azimuths=[-90,90] #^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
    az_offsets = [180]  #^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
    az_start = azimuths[0]
    
    
    el_speed  = 1
    elevations = [20,20]  #############[20, 85]
    el_offsets = [65] #^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
    el_start = elevations[0]  # [20, 85]^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
     
    
    
    wait5 = 5
    wait10 = 10
    wait100 = 100
    sampling = 100
    scu.debug = False
   
    # Activate the servo system
    print("Manual move to az: {} and el: {}".format(az_start, el_start))
    scu.activate_dmc()
###    scu.wait_duration(wait10*3)
    scu.wait_duration(wait10*2)

    # Command dish to starting position
    #az_angle = az_start
    #scu.abs_azimuth(az_angle, az_speed)
    #el_angle = el_start
    #scu.abs_elevation(el_angle, el_speed)
    #scu.wait_duration(wait10)

    for elev,azim in zip(elevations,azimuths):
        el_angle = elev
        az_angle = azim
        scu.abs_elevation(el_angle, el_speed)
        scu.wait_duration(1)
        scu.abs_azimuth(az_angle, az_speed)
        scu.wait_duration(150)    
        
        # Start ACU recording on SCU
        if scu.logger_state() == 'RECORDING':
            print('WARNING, already recording - attempting to stop and start a fresh logger')
            scu.stop_logger()  
            scu.wait_duration(wait10)
        if scu.logger_state() == 'STOPPED':
            print('Starting logger with config: {}'.format(config_name))
            scu.start_logger(config_name)
            scu.wait_duration(wait10)    

        for el_offset,az_offset in zip(el_offsets,az_offsets):
            print("\t El offset: {} ".format(el_offset))

            i=0
            while i < loop1:
                i+=1
                print("\t Loop 1 cycle: {} of {}".format(i, loop1))

                # Command Elevation with +offset and wait for 30 seconds            
                #el_angle = el_start + el_offset 
                scu.abs_elevation(el_angle + el_offset, el_speed)
                scu.wait_duration(1)
                scu.abs_azimuth(az_angle + az_offset, az_speed)
                scu.wait_duration(tcycle)

                # Command Elevation back to starting position and wait for 30 seconds
                #el_angle = el_start  
                scu.abs_elevation(el_angle, el_speed)
                scu.wait_duration(1)
                scu.abs_azimuth(az_angle, az_speed)
    ###            scu.wait_duration(wait10*6)            
                scu.wait_duration(tcycle)

                # Command Elevation with -offset and wait for 30 seconds            
                #el_angle = el_start - el_offset 
                #scu.abs_elevation(el_angle - el_offset, el_speed)
                #scu.wait_duration(wait10*2)

                # Command Elevation back to starting position and wait for 30 seconds
                #el_angle = el_start  
                #scu.abs_elevation(el_angle, el_speed)
    ###            scu.wait_duration(wait10*6)
                #scu.wait_duration(wait10*3)

            # Repeat above steps for step sizes of 1° and 10° at slew rate of 1°/sec             

        # Stop the ACU data recording
        scu.stop_logger()
        scu.wait_duration(wait10)
        print(scu.logger_state())

        # Extract the data on the SCU @ 5Hz sampling
        # Rename the files to TBD (eg. 6p8p6_i.csv)
        print("Rename the file to TBD (eg. 6p8p6_i.csv)")
        scu.save_session('6p12p4'+ '_az_' + str(azim) + '_el_' + str(elev), interval_ms = sampling)    
     
    # De-activate all axis
    scu.deactivate_dmc()
    scu.wait_duration(wait10*2) 
    
    # Manually stop the LDS data recording
    print ('Manually stop the LDS data recording')
    scu.wait_duration(wait10)
    
    print('Done')
```

```python
mpi_t686(mpi)
#mpi_t686(sim)
```
